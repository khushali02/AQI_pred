```python
# Importing necessary libraries

%pip install pandas
%pip install matplotlib
%pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('air_pollution_data.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
display(df.head())
```

```python
# Summary statistics
print("Summary statistics:")
display(df.describe())
```

```python
# Data types and missing values
print("Data types and missing values:")
display(df.info())
```

```python
# Correlation matrix
numeric_df = df.select_dtypes(include=['int64', 'float64'])
print("Correlation matrix:")
display(numeric_df.corr())
```

```python
# Histograms of numeric variables
print("Histograms of numeric variables:")
df.hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
# Box plots of numeric variables
print("Box plots of numeric variables:")
df.boxplot(figsize=(20, 10))
plt.xticks(rotation=0)
plt.ylim(0, 700)
plt.tight_layout()
plt.show()
```

```python
except_co_df = df.drop(columns=['co'])
print("Box plots of numeric variables:")
except_co_df.boxplot(figsize=(20, 10))
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
# Pairplot of numeric variables
print("Pairplot of numeric variables:")
sns.pairplot(df)
plt.show()
```
